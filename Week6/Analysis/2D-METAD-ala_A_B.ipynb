```python
# import modules
import sys
import numpy as np
import matplotlib.pyplot as plt

# read in data from fes.dat (skip first 9 lines)
data = np.loadtxt('/home/ac7407/comp-class/comp-lab-class/Week6/masterclass-21-4/masterclass-21-4/data/diala/ala_A/2D-METAD/fes.dat', skiprows=9)

# read in data before reshaping
phi = data[:,0]
psi = data[:,1]
fes = data[:,2]

# reshape data 
# phi_new is first 33 rows of phi
phi_new = phi[:33]

# psi_new is every 33 rows of psi
psi_new = psi[::33]

# fes_new needs to be reshaped to 18x33 with numpy.reshape
fes_new = np.reshape(fes, (18,33))

# plot data with imshow
plt.imshow(fes_new, origin='lower', extent=[phi_new[0], phi_new[-1], psi_new[0], psi_new[-1]], aspect='auto')
plt.colorbar()
plt.xlabel('phi')
plt.ylabel('psi')
plt.title("Free Energy Surface of Ala_A (Bias Factor = 6)")
plt.show()
```
