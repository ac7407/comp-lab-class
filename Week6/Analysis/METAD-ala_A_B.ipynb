```python
# read in data from fes.dat for ala_A starting from line 6
data_5 = np.genfromtxt('/home/ac7407/comp-class/comp-lab-class/Week6/masterclass-21-4/masterclass-21-4/data/diala/ala_A/fes.dat', skip_header=5)

# plot free energy as a function of phi using columns 0 and 1
plt.plot(data_5[:,0], data_5[:,1])
plt.xlabel('Phi (degrees)')
plt.ylabel('Free Energy (kJ/mol)')
plt.title('Free Energy as a function of Phi from a 10ns metadynamics simulation of ala_A in vacuum')
plt.show()

# files used are fes_1.dat, fes_3.dat, fes_10.dat, fes_20.dat, fes_50.dat, fes_100.dat
fes_1a = np.genfromtxt('/home/ac7407/comp-class/comp-lab-class/Week6/masterclass-21-4/masterclass-21-4/data/diala/ala_A/fes_1.dat', skip_header=5)
fes_3a = np.genfromtxt('/home/ac7407/comp-class/comp-lab-class/Week6/masterclass-21-4/masterclass-21-4/data/diala/ala_A/fes_3.dat', skip_header=5)
fes_10a = np.genfromtxt('/home/ac7407/comp-class/comp-lab-class/Week6/masterclass-21-4/masterclass-21-4/data/diala/ala_A/fes_10.dat', skip_header=5)
fes_20a = np.genfromtxt('/home/ac7407/comp-class/comp-lab-class/Week6/masterclass-21-4/masterclass-21-4/data/diala/ala_A/fes_20.dat', skip_header=5)
fes_50a = np.genfromtxt('/home/ac7407/comp-class/comp-lab-class/Week6/masterclass-21-4/masterclass-21-4/data/diala/ala_A/fes_50.dat', skip_header=5)
fes_100a = np.genfromtxt('/home/ac7407/comp-class/comp-lab-class/Week6/masterclass-21-4/masterclass-21-4/data/diala/ala_A/fes_100.dat', skip_header=5)

# plot free energy as a function of phi using columns 0 and 1
plt.plot(fes_1a[:,0], fes_1a[:,1], label='0.1ns', color='red')
plt.plot(fes_3a[:,0], fes_3a[:,1], label='0.3ns', color='green')
plt.plot(fes_10a[:,0], fes_10a[:,1], label='1ns', color='blue')
plt.plot(fes_20a[:,0], fes_20a[:,1], label='2ns', color='purple')
plt.plot(fes_50a[:,0], fes_50a[:,1], label='5ns', color='orange')
plt.plot(fes_100a[:,0], fes_100a[:,1], label='10ns', color='black')
plt.xlabel('Phi (degrees)')
plt.ylabel('Free Energy (kJ/mol)')
plt.title('Estimates of free energy as a function of Phi calculated every 100 Gaussian kernels deposited (ala_A)')
plt.legend()
plt.show()
```
