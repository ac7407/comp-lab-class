```python
# import modules
import numpy as np
import matplotlib.pyplot as plt
```

```python
# read potential energy vs time for QM/MM run using energy file
pot1 = np.loadtxt('/home/ac7407/comp-class/comp-lab-class/Week13/output/metad1_initmonitor/METADYN-1.ener')

# plot potential energy vs time for QM/MM using columns 0 and 4
plt.plot(pot1[:,0], pot1[:,4])
plt.xlabel('Time (fs)')
plt.ylabel('Potential Energy (au)')
plt.title('Potential Energy vs Time for QM/MM Run')
plt.show()
```
