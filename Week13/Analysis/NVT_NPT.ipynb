```python
# import modules
import numpy as np
import matplotlib.pyplot as plt

# read NVT/NPT energy file from output directory
NVT_1 = np.loadtxt('/home/ac7407/comp-class/comp-lab-class/Week13/output/NVT-1.ener')
NPT_1 = np.loadtxt('/home/ac7407/comp-class/comp-lab-class/Week13/output/NPT-1.ener')

# plot NVT/NPT energy file using columns 0 and 4
plt.plot(NVT_1[:,0], NVT_1[:,4], label='NVT')
plt.xlabel('Time (fs)')
plt.ylabel('Potential Energy (au)')
plt.legend()
plt.title('Potential Energy vs Time (NVT)')
plt.show()

# plot NPT energy file using columns 0 and 4
plt.plot(NPT_1[:,0], NPT_1[:,4], label='NPT')
plt.xlabel('Time (fs)')
plt.ylabel('Potential Energy (au)')
plt.legend()
plt.title('Potential Energy vs Time (NPT)')
plt.show()

```
