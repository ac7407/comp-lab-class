```python
import numpy as np
import matplotlib.pyplot as plt
import mdtraj as md

# load xtc file
traj = md.load_xtc('fit_trp-cage.xtc', top='step5_9.gro')

# calculate distance between N-terminal nitrogen and c-terminal alpha carbon
dist = md.compute_distances(traj, [[0, 9]])

# plot end-to-end distance over time
plt.plot(dist)
plt.xlabel('time (ps)')
plt.ylabel('distance (nm)')
plt.title('end to end distance between N-terminal a-nitrogen and C-terminal a-carbon vs time (ps)')
plt.show()
```

```python
# plot radius of gyration vs time
plt.plot(traj.time, md.compute_rg(traj))
plt.xlabel('Time (ps)')
plt.ylabel('Radius of gyration (nm)')
plt.title('Radius of gyration (nm) vs time (ps)')
plt.show()
```
