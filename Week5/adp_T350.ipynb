```python
import mdtraj as md
import matplotlib.pyplot as plt
import numpy as np

# load trajectory file
traj = md.load('adp_exchange4temps.trr', top='adp_exchange4temps.gro')

# compute phi and psi dihedral angles for each frame
phi = md.compute_phi(traj)
psi = md.compute_psi(traj)

# create 2d histogram of psi and phi angles
# use 100 bins in each dimension
# set maximum free energy to 10 kT
plt.hist2d(phi[1][:,0], psi[1][:,0], bins=100, cmap='Purples', vmax=10)
plt.xlabel('phi')
plt.ylabel('psi')
plt.title("350K")
plt.colorbar()
plt.show()
```
