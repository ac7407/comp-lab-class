```python
# import modules
import mdtraj as md
import numpy as np
```

```python
# load trajectory file
traj = md.load('adp.trr', top='adp.gro')

# compute psi and phi dihedral angles for each frame
psi = md.compute_phi(traj)
phi = md.compute_psi(traj)
```

```python
# plot psi angle vs time (in ns)
import matplotlib.pyplot as plt
plt.plot(traj.time/1000, psi[1], color='blue')
plt.xlabel('Time (ns)')
plt.ylabel('Psi angle (rad)')
plt.show()

# plot phi angle vs time (in ns)
plt.plot(traj.time/1000, phi[1], color='red')
plt.xlabel('Time (ns)')
plt.ylabel('Phi angle (rad)')
plt.show()
```

```python
# create 2d histogram of psi and phi angles
# use 100 bins in each dimension
plt.hist2d(psi[1][:,0], phi[1][:,0], bins=100, cmap='Purples')
plt.xlabel('Psi angle (rad)')
plt.ylabel('Phi angle (rad)')
plt.title('300K')
plt.colorbar()
plt.show()
```
